```python
#DATA PREPROCESSING
```

```python
import datetime
```

```python
date_time = datetime.datetime.now()
```

```python
print(date_time)
```

```python
df['Age']=date_time.year - df['Year']
```

```python
df.head(5)
```

```python
df.drop('Year',axis=1,inplace=True)
```

```python
df.head()
```

```python
#outlier removal
```

```python
sns.boxplot(df['Selling_Price'])
```

```python
sorted(df['Selling_Price'],reverse=True) #as two points are outliers 33.0 and 35.0
```

```python
df[(df['Selling_Price']>=33.0)&(df['Selling_Price']<=35.0)]
```

```python
df = df[~(df['Selling_Price']>=33.0)&(df['Selling_Price']<=35.0)]
```

```python
df.shape
```

```python
#encoding the categorial columns
```

```python
df['Fuel_Type'].unique()
```

```python
df['Fuel_Type'] = df['Fuel_Type'].map({'Petrol':0, 'Diesel':1, 'CNG':2})
```

```python
df['Fuel_Type'].unique()
```

```python
df['Selling_type'].unique()
```

```python
df['Selling_type'] = df['Selling_type'].map({'Dealer':0, 'Individual':1})
```

```python
df['Selling_type'].unique()
```

```python
df['Transmission'].unique()
```

```python
df['Transmission'] = df['Transmission'].map({'Manual':0, 'Automatic':1})
```

```python
df['Transmission'].unique()
```

```python
df.head()
```

```python
#store feature matrix in x and response(target) in vector y
```

```python
x = df.drop(['Car_Name','Selling_Price'],axis=1)
y = df['Selling_Price']
```

```python
x
```

```python
y
```

```python
#splitting dataset in test train split
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
#import the models

#as selling price is the out put column has continous values then it is regression problem
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
#model training
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
rf = RandomForestRegressor()
rf.fit(x_train,y_train)
```

```python
gbr = GradientBoostingRegressor()
gbr.fit(x_train,y_train)
```

```python
xg = XGBRegressor()
xg.fit(x_train,y_train)
```

```python
#prediction on test data
```

```python
y_pred1 = lr.predict(x_test)
y_pred2 = rf.predict(x_test)
y_pred3 = gbr.predict(x_test)
y_pred4 = xg.predict(x_test)
```

```python
#evaluating the algorithm
```

```python
from sklearn import metrics
```

```python
score1 = metrics.r2_score(y_test,y_pred1)
score2 = metrics.r2_score(y_test,y_pred2)
score3 = metrics.r2_score(y_test,y_pred3)
score4 = metrics.r2_score(y_test,y_pred4)
```

```python
print(score1,score2,score3,score4)
```

```python
final_data = pd.DataFrame({'Models':['lr','rf','gbr','xg'],"R2_SCORE":[score1,score2,score3,score4]})
```

```python
final_data
```

```python
sns.barplot(x='Models', y='R2_SCORE', data=final_data)
```
