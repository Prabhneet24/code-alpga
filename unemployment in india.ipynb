```python
df=  df.fillna(df.mean)
```

```python
df = df.rename(columns={'Estimated Unemployment Rate (%) ': 'Estimated Unemployment Rate (%)'})
```

```python
print(df.columns)
```

```python
df[' Estimated Unemployment Rate (%)'] = pd.to_numeric(df[' Estimated Unemployment Rate (%)'], errors='coerce')
```

```python
regional_unemployment = df.groupby('Region')[' Estimated Unemployment Rate (%)'].mean()
```

```python
regional_unemployment
```

```python
df[' Estimated Labour Participation Rate (%)'] = pd.to_numeric(df[' Estimated Labour Participation Rate (%)'], errors='coerce')
```

```python
labour_participation = df.groupby('Region')[' Estimated Labour Participation Rate (%)'].mean()
```

```python
labour_participation
```

```python
df[' Estimated Employed']= pd.to_numeric(df[' Estimated Employed'],errors='coerce')
```

```python
region_wise_employed = df.groupby('Region')[' Estimated Employed'].mean()
```

```python
region_wise_employed
```

```python
import matplotlib.pyplot as plt
plt.hist(df[' Estimated Unemployment Rate (%)'], bins=50)
plt.xlabel(' Estimated Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.title('Distribution of Unemployment Rates')
plt.show()
```

```python


plt.scatter(df[' Estimated Unemployment Rate (%)'], df[' Estimated Employed'])
plt.xlabel(' Estimated Unemployment Rate (%)')
plt.ylabel(' Estimated Employed')
plt.title('Relationship between Unemployment Rate and Employed')
plt.show()

```

```python
plt.scatter(df[' Estimated Unemployment Rate (%)'], df[' Estimated Labour Participation Rate (%)'])
plt.xlabel(' Estimated Unemployment Rate (%)')
plt.ylabel(' Estimated Labour Participation Rate (%)')
plt.title('Relationship between Unemployment Rate and Labour Participation Rate')
plt.show()
```
