```python
plt.hist(data['Sales'], bins=5)
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.title('Sales Distribution')
plt.show()
```

```python
corr_matrix = data.corr()

print(corr_matrix)
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

axs[0, 0].hist(data['TV'], bins=5)
axs[0, 0].set_title('TV Distribution')

axs[0, 1].hist(data['Radio'], bins=5)
axs[0, 1].set_title('Radio Distribution')

axs[0, 2].hist(data['Newspaper'], bins=5)
axs[0, 2].set_title('Newspaper Distribution')

axs[1, 0].hist(data['Sales'], bins=5)
axs[1, 0].set_title('Sales Distribution')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
tv_sales = data['TV'].sum()
radio_sales = data['Radio'].sum()
newspaper_sales = data['Newspaper'].sum()

```

```python
# Create a pie chart of the sales
plt.figure(figsize=(10, 8))
plt.pie([tv_sales, radio_sales, newspaper_sales], labels=['TV', 'Radio', 'Newspaper'], autopct='%1.1f%%')
plt.title('Pie Chart of Sales by Advertising Type')
plt.show()
```
